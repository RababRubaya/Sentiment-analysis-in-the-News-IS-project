```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
from bs4 import BeautifulSoup 
import sys 


def review_to_words( raw_review ):
    review_text = BeautifulSoup(raw_review, "lxml").get_text() 
    letters_only = re.sub("[^a-zA-Z]", " ", review_text) 
    words = letters_only.lower().split()
    ps = PorterStemmer() 
    r = open("stemming.txt", "w")
    for w in words: 
        if(not(w == ps.stem(w))):
            #print(w, ps.stem(w))
            entry = w + "," + ps.stem(w) + "\n"
            r.write(entry)
    stops = set(stopwords.words("english"))                  
    meaningful_words = [w for w in words if not w in stops]
    uniqueWords = [ps.stem(word) for word in meaningful_words]
    return(" ".join( uniqueWords))   
```

```python
train = pd.read_csv("trainData.tsv", encoding = "cp1252", header=0, \
                    delimiter="\t")
clean_train_reviews = []
for i in range( 0, train["contents"].size ):
    clean_train_reviews.append( review_to_words( train["contents"][i]))
    
    if( (i+1)%1000 == 0):
        print("Review %d of %d\n" % ( i+1, train["contents"].size ))              
    
print("Creating the bag of words...\n")
vectorizer = CountVectorizer(analyzer = "word",   \
                             tokenizer = None,    \
                             preprocessor = None, \
                             stop_words = None,   \
                             max_features = 5000) #change to 5000

data_features = vectorizer.fit_transform(clean_train_reviews)
data_features = data_features.toarray()
vocab = vectorizer.get_feature_names()
#print(vocab)
dist = np.sum(data_features, axis=0)

f = open("bagOfWords.txt", "w")
for tag, count in zip(vocab, dist):
    entry = tag + "," + str(count) + "\n"
    print(tag, count)
    f.write(entry)
f.close()
```
